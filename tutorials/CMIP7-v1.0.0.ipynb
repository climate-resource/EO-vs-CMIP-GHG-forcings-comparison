```python
# imports
from src.utils_preprocessing import prepare_data
from src.utils_plotting import plot_data
import pandas as pd
```

## Load data

```python
df_co2 = pd.read_csv("../data/CMIP7_CO2.csv")
df_ch4 = pd.read_csv("../data/CMIP7_CH4.csv")
```

## Conditions

+ select a grid size via `grid_deg`. Valid values are `15, 30, 45, 90`.
+ select a specific latitudes via `deg_only` (can be one or more values in a list)
+ specify custom range of years via `year_min`, `year_max`


```python
# select northern and southern hemisphere only
df_prep = prepare_data(df_co2, grid_deg=90, ghg_gas="co2", year_min=2000)
plot_data(df_prep, ghg_gas="co2", global_mean=True)
```

```python
# select  30-degrees grid
df_prep = prepare_data(df_co2, grid_deg=30, ghg_gas="co2", year_min=1950, year_max=2000)
plot_data(df_prep, ghg_gas="co2", global_mean=True)
```

```python
# select only 15 and 30 degree latitude
df_prep = prepare_data(df_ch4, deg_only=[15, 30], ghg_gas="ch4", year_min=1950, year_max=2000)
plot_data(df_prep, ghg_gas="ch4", global_mean=False)
```
